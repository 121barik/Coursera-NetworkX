# Network Data Science with NetworkX and Python

## Generate random graphs

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
graph = nx.complete_graph(10)
```

```python
nx.draw(graph, 
        node_color='C1',
        with_labels=True)
```

```python
graph = nx.binomial_tree(5)
nx.draw(graph, 
        node_color='C1',
        with_labels=True)
```

```python
graph = nx.binomial_graph(20, 0.15)
nx.draw(graph, 
        node_color='C1',
        with_labels=True)
```

```python
graph = nx.connected_caveman_graph(5, 10)
nx.draw(graph, 
        node_color='C1',
        with_labels=True)
```

```python
graph = nx.random_internet_as_graph(100)
nx.draw(graph, 
        node_color='C1',
        with_labels=True)
```

```python
plt.figure(figsize=(10, 15))
graph = nx.les_miserables_graph()
nx.draw(graph, 
        node_color='C1',
        with_labels=True)
```
